```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Data
df=pd.DataFrame({
"Budget":[1000,2000,1500,3000,1200],
"Ads":[500,800,600,1200,400],
"Footfall":[150,280,210,420,180]
})

X=df[["Budget","Ads"]]
y=df["Footfall"]

# Q1: Scatter Plot
plt.scatter(df["Budget"],y)
plt.show()

# Q2: Train Model
model=LinearRegression().fit(X,y)

# Q3: Predict Footfall
print(model.predict([[2500,900]]))

# Q4: Coefficients
print(model.coef_)

# Q5: Pair Plot
pd.plotting.scatter_matrix(df)
plt.show()

# Q6: Coefficient Bar Chart
plt.bar(["Budget","Ads"],model.coef_)
plt.show()
```
