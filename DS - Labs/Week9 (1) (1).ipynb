```python
# Probability Distributions using Python

# Application: Student Exam Pass/Fail Analysis
import numpy as np
from scipy.stats import bernoulli, binom

# BERNOULLI DISTRIBUTION
print("----- Bernoulli Distribution -----")

p = 0.7
bernoulli_samples = bernoulli.rvs(p, size=10)
print("Probability of Passing:", p)
print("Student Outcomes (1=Pass, 0=Fail):")
print(bernoulli_samples)
# Probability of success and failure
print("\nProbability of Pass:")
print(bernoulli.pmf(1, p))
print("Probability of Fail:")
print(bernoulli.pmf(0, p))
# -----------------------------
# BINOMIAL DISTRIBUTION
# -----------------------------
print("\n\n----- Binomial Distribution -----")
# Number of students
n = 10

# Probability of passing
p = 0.7
# Probability that exactly 8 students pass
k = 8
binomial_probability = binom.pmf(k, n, p)
print("Number of Students:", n)
print("Probability of Passing:", p)
print(f"\nProbability that exactly {k} students pass:")
print(binomial_probability)
# Generate random experiment
random_binomial = binom.rvs(n, p, size=5)
print("\nRandom Binomial Outcomes:")
print(random_binomial)
```

```python
# Common Discrete Probability Distributions

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, geom, poisson, nbinom, randint

p = 0.6
x = np.array([0, 1])
pmf = bernoulli.pmf(x, p)
print("Possible Values:", x)
print("PMF:", pmf)
samples = bernoulli.rvs(p, size=20)
print("Random Samples:", samples)
plt.figure(figsize=(5,4))
plt.bar(x, pmf)
plt.title("Bernoulli Distribution")
plt.xlabel("X")
plt.ylabel("Probability")
plt.show()
# BINOMIAL DISTRIBUTION
n = 10
p = 0.5
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
print("PMF:")
print(pmf)
samples = binom.rvs(n, p, size=20)
print("Random Samples:", samples)
plt.figure(figsize=(7,4))
plt.bar(x, pmf)
plt.title("Binomial Distribution")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
# GEOMETRIC DISTRIBUTION
p = 0.3
x = np.arange(1,11)
pmf = geom.pmf(x, p)
print("PMF:")
print(pmf)
samples = geom.rvs(p, size=20)
print("Random Samples:", samples)
plt.figure(figsize=(7,4))
plt.bar(x, pmf)
plt.title("Geometric Distribution")
plt.xlabel("Trial")
plt.ylabel("Probability")
plt.show()
# POISSON DISTRIBUTION
lam = 4
x = np.arange(0,15)
pmf = poisson.pmf(x, lam)
print("PMF:")
print(pmf)
samples = poisson.rvs(lam, size=20)
print("Random Samples:", samples)
plt.figure(figsize=(7,4))
plt.bar(x, pmf)
plt.title("Poisson Distribution")
plt.xlabel("Events")
plt.ylabel("Probability")
plt.show()
# NEGATIVE BINOMIAL DISTRIBUTION
r = 5
p = 0.4
x = np.arange(0,15)
pmf = nbinom.pmf(x, r, p)
print("PMF:")
print(pmf)
samples = nbinom.rvs(r, p, size=20)
print("Random Samples:", samples)
plt.figure(figsize=(7,4))
plt.bar(x, pmf)
plt.title("Negative Binomial Distribution")
plt.xlabel("Failures")
plt.ylabel("Probability")
plt.show()
# DISCRETE UNIFORM DISTRIBUTION
low = 1
high = 7        # Generates numbers 1 to 6
x = np.arange(low, high)
pmf = randint.pmf(x, low, high)
print("PMF:")
print(pmf)
samples = randint.rvs(low, high, size=20)
print("Random Samples:", samples)
plt.figure(figsize=(6,4))
plt.bar(x, pmf)
plt.title("Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()
```

```python
# Common Continuous Probability Distributions

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import (
    norm,
    uniform,
    expon,
    gamma,
    beta,
    lognorm,
    chi2,
    t,
    weibull_min
)
# 1. NORMAL DISTRIBUTION
mu = 50
sigma = 10
x = np.linspace(20, 80, 200)
pdf = norm.pdf(x, mu, sigma)
samples = norm.rvs(mu, sigma, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Normal Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 2. UNIFORM DISTRIBUTION
a = 0
b = 10
x = np.linspace(a, b, 200)
pdf = uniform.pdf(x, a, b)
samples = uniform.rvs(a, b, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Uniform Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 3. EXPONENTIAL DISTRIBUTION
print("\n========== Exponential Distribution ==========")
scale = 2
x = np.linspace(0, 15, 200)
pdf = expon.pdf(x, scale=scale)
samples = expon.rvs(scale=scale, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Exponential Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 4. GAMMA DISTRIBUTION
shape = 2
x = np.linspace(0, 15, 200)
pdf = gamma.pdf(x, a=shape)
samples = gamma.rvs(a=shape, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Gamma Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 5. BETA DISTRIBUTION
print("\n========== Beta Distribution ==========")
a = 2
b = 5
x = np.linspace(0,1,200)
pdf = beta.pdf(x, a, b)
samples = beta.rvs(a, b, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Beta Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 6. LOGNORMAL DISTRIBUTION
sigma = 0.5
x = np.linspace(0.01,5,200)
pdf = lognorm.pdf(x, sigma)
samples = lognorm.rvs(sigma, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Log-Normal Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 7. CHI-SQUARE DISTRIBUTION
df = 5
x = np.linspace(0,20,200)
pdf = chi2.pdf(x, df)
samples = chi2.rvs(df, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Chi-Square Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 8. STUDENT'S T DISTRIBUTION
df = 10
x = np.linspace(-5,5,200)
pdf = t.pdf(x, df)
samples = t.rvs(df, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Student's t Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
# 9. WEIBULL DISTRIBUTION
shape = 2
x = np.linspace(0,5,200)
pdf = weibull_min.pdf(x, shape)
samples = weibull_min.rvs(shape, size=20)
print("Random Samples:")
print(samples)
plt.figure(figsize=(7,4))
plt.plot(x, pdf)
plt.title("Weibull Distribution")
plt.xlabel("X")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
# Problem statement 1

import numpy as np
from scipy.stats import norm
# Step 1: Generate random population of 1000 student marks between 50 and 95
np.random.seed(42)
population = np.random.randint(50, 96, 1000)
# Step 2: Select 10% sample from population
sample_size = int(0.10 * len(population))
sample = np.random.choice(population, size=sample_size, replace=False)
# Step 3: Calculate population mean and population standard deviation
population_mean = np.mean(population)
population_std = np.std(population)
# Step 4: Calculate sample mean
sample_mean = np.mean(sample)
# Step 5: Calculate Z-test statistic
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
# Step 6: Calculate p-value for two-tailed Z-test
p_value = 2 * (1 - norm.cdf(abs(z_score)))
# Step 7: Display results
print("Population Size:", len(population))
print("Sample Size:", sample_size)
print("Population Mean:", population_mean)
print("Population Standard Deviation:", population_std)
print("Sample Mean:", sample_mean)
print("Z-score:", z_score)
print("P-value:", p_value)
# Step 8: Decision
alpha = 0.05
if p_value < alpha:
    print("Decision: Reject the Null Hypothesis")
else:
    print("Decision: Fail to Reject the Null Hypothesis")
```

```python
# # Statistical Hypothesis Testing
# 1. Z-Test
# 2. Independent t-Test
# Application: Student Performance Analysis
import numpy as np
from scipy import stats
from statsmodels.stats.weightstats import ztest

# Z-TEST
print("----- Z-Test -----")
# Sample marks of students
sample_marks = np.array([78, 82, 85, 88, 90, 76, 95, 89, 84, 91])
# Population mean
population_mean = 80
# Perform one-sample Z-test
z_stat, p_value = ztest(sample_marks, value=population_mean)
print("Sample Marks:")
print(sample_marks)
print("\nPopulation Mean:", population_mean)
print("\nZ-Statistic:", z_stat)
print("P-Value:", p_value)
# Decision
alpha = 0.05
if p_value < alpha:
    print("\nResult: Reject Null Hypothesis")
else:
    print("\nResult: Fail to Reject Null Hypothesis")
# INDEPENDENT t-TEST
print("\n\n----- Independent t-Test -----")
# Marks of two different classes
class_A = np.array([78, 82, 85, 88, 90])
class_B = np.array([72, 75, 70, 68, 74])
# Perform independent t-test
t_stat, p_value = stats.ttest_ind(class_A, class_B)
print("Class A Marks:")
print(class_A)
print("\nClass B Marks:")
print(class_B)
print("\nt-Statistic:", t_stat)
print("P-Value:", p_value)
# Decision
if p_value < alpha:
    print("\nResult: Reject Null Hypothesis")
else:
    print("\nResult: Fail to Reject Null Hypothesis")


```

```python
# Parametric Statistical Tests

import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.stats.weightstats import ztest
from statsmodels.stats.proportion import proportions_ztest

# Sample Data
np.random.seed(42)
group1 = np.random.normal(loc=70, scale=10, size=30)
group2 = np.random.normal(loc=75, scale=10, size=30)
group3 = np.random.normal(loc=80, scale=10, size=30)
before = np.random.normal(loc=65, scale=8, size=30)
after = before + np.random.normal(loc=5, scale=4, size=30)

# 1. ONE-SAMPLE Z-TEST
z_stat, p_value = ztest(group1, value=70)
print("Z-statistic:", z_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")
# 2. TWO-SAMPLE Z-TEST
z_stat, p_value = ztest(group1, group2)
print("Z-statistic:", z_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")
# 3. ONE-PROPORTION Z-TEST
success_count = 60
sample_size = 100
claimed_proportion = 0.50
z_stat, p_value = proportions_ztest(
    count=success_count,
    nobs=sample_size,
    value=claimed_proportion
)
print("Z-statistic:", z_stat)
print("P-value:", p_value)
# 4. TWO-PROPORTION Z-TEST
successes = np.array([60, 75])
samples = np.array([100, 120])
z_stat, p_value = proportions_ztest(
    count=successes,
    nobs=samples
)
print("Z-statistic:", z_stat)
print("P-value:", p_value)
# 5. ONE-SAMPLE T-TEST
t_stat, p_value = stats.ttest_1samp(group1, popmean=70)
print("T-statistic:", t_stat)
print("P-value:", p_value)
# 6. INDEPENDENT TWO-SAMPLE T-TEST
t_stat, p_value = stats.ttest_ind(group1, group2)
print("T-statistic:", t_stat)
print("P-value:", p_value)
# 7. PAIRED SAMPLE T-TEST
t_stat, p_value = stats.ttest_rel(before, after)
print("T-statistic:", t_stat)
print("P-value:", p_value)
# 8. ONE-WAY ANOVA
f_stat, p_value = stats.f_oneway(group1, group2, group3)
print("F-statistic:", f_stat)
print("P-value:", p_value)
# 9. PEARSON CORRELATION TEST
x = np.random.normal(50, 10, 30)
y = x + np.random.normal(5, 5, 30)
r_value, p_value = stats.pearsonr(x, y)
print("Correlation Coefficient:", r_value)
print("P-value:", p_value)
# 10. SIMPLE LINEAR REGRESSION TEST
slope, intercept, r_value, p_value, std_error = stats.linregress(x, y)

print("Slope:", slope)
print("Intercept:", intercept)
print("R-value:", r_value)
print("P-value:", p_value)
print("Standard Error:", std_error)
# 11. F-TEST FOR VARIANCE
var1 = np.var(group1, ddof=1)
var2 = np.var(group2, ddof=1)
f_stat = var1 / var2
df1 = len(group1) - 1
df2 = len(group2) - 1
p_value = 2 * min(
    stats.f.cdf(f_stat, df1, df2),
    1 - stats.f.cdf(f_stat, df1, df2)
)
print("F-statistic:", f_stat)
print("P-value:", p_value)
```

```python
#  Python Program: Non-Parametric Statistical Tests

import numpy as np
from scipy import stats
np.random.seed(42)

group1 = np.array([45, 50, 52, 55, 58, 60, 62, 65])
group2 = np.array([48, 53, 57, 59, 61, 64, 67, 70])
group3 = np.array([40, 44, 49, 51, 54, 56, 59, 63])

before = np.array([50, 52, 55, 57, 60, 62, 64, 66])
after  = np.array([53, 55, 56, 60, 63, 65, 67, 70])

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([2, 4, 5, 7, 8, 10, 11, 13])

observed = np.array([25, 30, 45])
expected = np.array([33.33, 33.33, 33.34])

u_stat, p_value = stats.mannwhitneyu(group1, group2)

print("U-statistic:", u_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")

w_stat, p_value = stats.wilcoxon(before, after)

print("W-statistic:", w_stat)
print("P-value:", p_value)


print("\n========== Kruskal-Wallis H Test ==========")

h_stat, p_value = stats.kruskal(group1, group2, group3)

print("H-statistic:", h_stat)
print("P-value:", p_value)


print("\n========== Friedman Test ==========")

condition1 = np.array([55, 60, 62, 64, 66])
condition2 = np.array([58, 61, 65, 67, 69])
condition3 = np.array([60, 63, 66, 70, 72])

friedman_stat, p_value = stats.friedmanchisquare(
    condition1, condition2, condition3
)

print("Friedman Statistic:", friedman_stat)
print("P-value:", p_value)

rho, p_value = stats.spearmanr(x, y)

print("Spearman Correlation:", rho)
print("P-value:", p_value)


tau, p_value = stats.kendalltau(x, y)

print("Kendall Tau:", tau)
print("P-value:", p_value)

print("\n========== Chi-Square Goodness of Fit Test ==========")

chi_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)


table = np.array([
    [30, 10],
    [20, 40]
])

chi_stat, p_value, dof, expected_freq = stats.chi2_contingency(table)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:")
print(expected_freq)

sample = np.random.uniform(0, 1, 100)

ks_stat, p_value = stats.kstest(sample, 'uniform')

print("KS Statistic:", ks_stat)
print("P-value:", p_value)
stat, p_value, median, contingency_table = stats.median_test(
    group1, group2, group3
)

print("Statistic:", stat)
print("P-value:", p_value)
print("Grand Median:", median)
print("Contingency Table:")
print(contingency_table)
differences = after - before

positive_signs = np.sum(differences > 0)
negative_signs = np.sum(differences < 0)

n = positive_signs + negative_signs

p_value = stats.binomtest(
    positive_signs,
    n,
    p=0.5,
    alternative='two-sided'
).pvalue

print("Positive Signs:", positive_signs)
print("Negative Signs:", negative_signs)
print("P-value:", p_value)

```
