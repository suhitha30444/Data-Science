Central Tendency

```python
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()

df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print("Mean")
print(df.mean())

print("Median")
print(df.median())

print("Mode")
print(df.mode().iloc[0])
```

Dispersion Measures

```python
print("Range")
print(df.max() - df.min())

print("Variance")
print(df.var())

print("Standard Deviation")
print(df.std())

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

IQR = Q3 - Q1

print("IQR")
print(IQR)
```

Skewness & Kurtosis

```python
print("Skewness")
print(df.skew())

print("Kurtosis")
print(df.kurt())
```
